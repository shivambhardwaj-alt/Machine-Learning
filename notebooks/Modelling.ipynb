Training the Model 

Linear Regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
from sklearn.linear_model import LassoLars
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('./data/preprocessed.csv')
```

```python
print(df.head())
```

```python
Y = df.Sales
X = df.drop(columns = ['Sales'],axis = 1)
print(Y.head())
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X , Y , test_size=0.3,random_state=42)


# print(X_train.isnull().sum())
print(np.isinf(X_train).sum().sum())




```

```python
regressor = LinearRegression()
cv_score = cross_val_score(regressor,X_train,Y_train,cv = 5 ,scoring='r2')
print(cv_score)
print('Average CV_Score is : ', np.mean(cv_score))
```

```python
y_pred = regressor.predict(X_test)
mse = mean_squared_error(Y_test,y_pred)
print('Mean_squared_erorr is : ' ,mse)
print('Root mean Squared error is : ' , np.sqrt(mse))
print(r2_score(Y_test,y_pred))
```

Lasso Regression

```python
lasso = LassoLars(alpha=0.1, fit_intercept=False)

#trying cross validation for this
cv_scores = cross_val_score(lasso, X_train, Y_train, cv=5, scoring='r2')

print("Cross-validation R² scores:", cv_scores)
print("Average R² score:", np.mean(cv_scores))
```

DecisionTreeRegression

```python
param_grid = {
    'max_depth': [3, 5, 10, None],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4], 
    'max_features': [None, 'sqrt', 'log2'],  
    'criterion': ['squared_error', 'friedman_mse', 'poisson'],  
}
```

```python
tree = DecisionTreeRegressor(random_state=42)
grid_search = GridSearchCV(estimator=tree,param_grid=param_grid,cv = 5,scoring='neg_mean_squared_error',n_jobs=1,verbose=1)
grid_search.fit(X_train,Y_train)
```
